```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("books.csv")
df.head()
```

```python
df = df.fillna("")

df["content"] = (
    df["genre"] + " " +
    df["author"] + " " +
    df["description"]
)
```

```python
tfidf = TfidfVectorizer(stop_words="english")

tfidf_matrix = tfidf.fit_transform(df["content"])
```

```python
similarity = cosine_similarity(tfidf_matrix)
```

```python
indices = pd.Series(df.index, index=df["title"]).drop_duplicates()

def recommend(title, top_n=5):

    if title not in indices:
        print(f"The title '{title}' was not found in the dataset. Please provide a valid book title from the list.")
        return

    idx = indices[title]

    scores = list(enumerate(similarity[idx]))

    scores = sorted(scores, key=lambda x:x[1], reverse=True)

    scores = scores[1:top_n+1]

    book_index = [i[0] for i in scores]

    return df[["title","author","genre"]].iloc[book_index]
```

```python
recommend("Harry Potter and the Sorcerer's Stone")
```
